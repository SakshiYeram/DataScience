```python
import json
```

```python
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
    
```

```python
data = load_data("massive.json")
```

```python
data
```

```python
def display_data(data):
    print("User and their Connections: ")

    for user in data['users']:
        print(f"ID: {user['id']} {user['name']} is friend with : {user['friends']} and they liked pages {user['liked_pages']} ")
    print("pages information:")
    for page in data['pages']:
        print(f"ID: {page['id']} and Name: {page['name']} ")
display_data(data)
```
