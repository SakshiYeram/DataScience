```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]


kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years,kohli,label= "Virat kohli")
plt.plot(years,sehwag,label= "Virender Shewag")
plt.xlabel("Years")
plt.ylabel("Run Scored")
plt.title("Run Comparison")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(years,kohli,linestyle="-.",label= "Virat kohli")
plt.plot(years,sehwag,linestyle="--",label= "Virender Shewag")
plt.plot(years,runs,label = "Sachin Tendulkar")
plt.xlabel("Years")
plt.ylabel("Run Scored")
plt.title("Run Comparison")
plt.grid(True)
plt.legend()

plt.show()
```
