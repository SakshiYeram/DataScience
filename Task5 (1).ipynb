Plot a Line Chart: Plot y = 3x + 2 for x in the range 0 to 10. Set the x-axis and y-axis labels, 
and give the plot a title.

```python
x = np.arange(0,11,1)

y = 3*x + 2

plt.plot(x,y,marker = 'o',color="Blue",linestyle = "--")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line plot : y = 3x+2")
plt.grid(True)
plt.show()
```

Bar Chart with Annotations: Draw a vertical bar chart for five categories of your choice. 
Annotate the value on top of each bar. 

```python
city = ["Mumbai","Delhi","Bengaluru","Chennai","Kolkata"]
sales = [85,72,64,58,49]

plt.bar(city,sales,color="lightgreen",edgecolor="green")
plt.title(" Bar Chart of City and Sales")
plt.xlabel("City")
plt.ylabel("Sales")

for i, sale in enumerate(sales):
    plt.text(i,sale + 0.8,str(sales[i]),ha = 'center',fontsize=10)
plt.show()
```

Multiple Lines in One Plot: Plot three mathematical functions (sin, cos, exp) on the same 
axes, each with different colors and line styles. Add a legend. 

```python
x = np.arange(0,np.pi,0.1)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(np.array(x) / 2 )

plt.plot(y1,color="blue",linestyle="--",label="sinx")
plt.plot(y2,color="red",linestyle="-.",label="cosx")
plt.plot(y3,color="green",linestyle=":",label="exp(x/2)")
plt.title("Multiple function in one shot")
plt.xlabel("x values")
plt.ylabel("Function values")
plt.legend()
plt.grid(True)
plt.show()
```

Stack Plot: Plot the cumulative sales of 3 products over 12 months as a stack plot using 
made-up data.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

product_a = [30, 25, 35, 40, 45, 50, 55, 53, 60, 65, 70, 75]
product_b = [20, 22, 25, 28, 30, 35, 33, 37, 40, 42, 45, 50]
product_c = [15, 18, 20, 22, 25, 28, 27, 30, 32, 35, 37, 40]

labels = ["Product A","Product B","Product C"]
colors = ["skyblue","lightgreen", "pink"]

plt.stackplot(months,product_a,product_b,product_c,labels=labels,colors=colors)
plt.title("cumulative sales of 3 products over 12 months")
plt.xlabel("Products")
plt.ylabel("Cumulative sale")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
