1)Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.  
Then create a horizontal bar chart showing total runs scored in their debut 5 years.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [2003,2007,2010,2012,2015]
sachin = [500, 700, 1100, 1500, 1800]
virat = [0, 500, 800, 1100, 1300]
virender = [100, 300, 800, 1200, 1500]
```

```python
x = np.arange(len(years))
x
```

```python
x = np.arange(len(years))
width = 0.25

plt.bar(x - width, sachin, width=width, label="Sachin Tendulkar")
plt.bar(x, virat, width=width, label="Virat Kohli")
plt.bar(x + width, virender, width=width, label="Virender Sehwag")
plt.title("Run Comparison")
plt.xlabel("years")
plt.ylabel("Run scored")
plt.xticks(x,years)
plt.legend()
plt.show()
```

```python
players = ["sachin","virat","virender"]
runs = [500+700+1100+1500+1800 , 0+500+800+1100+1300 , 100+300+800+1200+1500]
plt.barh(players,runs,color="pink")
plt.title("Performance of Indian Batmen in 5 Years")
plt.xlabel("Total Runs in 5 Years")
plt.show()
```

2) Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same using a horizontal bar chart and observe which is easier to understand.

```python
plt.style.use('ggplot')
```

```python
labels = ["Andoid","iOS","others"]
Market_share =[71,28,1]
colors = ['#ff9999','#66b3ff','#99ff99']
explode = [0.1,0,0]
plt.pie(Market_share,labels=labels,startangle=90,colors=colors,explode=explode,shadow=True,autopct='%1.1f%%')
plt.title("Market Share of Mobile")
plt.show()


```
